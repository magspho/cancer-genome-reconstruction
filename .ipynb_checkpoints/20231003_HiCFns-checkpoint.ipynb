```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
plt.rcParams["font.family"] = "Arial"
```

```python
chr_bands = pd.read_csv('chromosome_banding_hg38.txt', sep='\t')
band_colors = []
for i in chr_bands['gieStain']:
    if i == 'acen':
        band_colors.append('Red')
    elif i == 'gneg':
        band_colors.append('White')
    elif i == 'gpos100':
        band_colors.append('Black')
    elif i == 'gpos25':
        band_colors.append('Silver')
    elif i == 'gpos50':
        band_colors.append('Gray')
    elif i == 'gpos75':
        band_colors.append('Gray')
    elif i == 'stalk':
        band_colors.append('None')
    elif i == 'gvar':
        band_colors.append('Blue')
chr_bands['color'] = band_colors

def getSize(chrom):
    chrom_sizes = pd.read_csv('hg38_chrom_sizes.txt', header=None, sep='\t')
    return int(chrom_sizes[chrom_sizes[0]==chrom][1])

def makeContact(links1, links2, chrom1, chrom2, binsize):
    pos1 = np.unique(links1[2])
    pos2 = np.unique(links2[2])
    bins1 = range(0,getSize(chrom1), binsize)
    bins2 = range(0,getSize(chrom2), binsize)
    bin_ids_1 = np.digitize(links1[2], bins1)
    bin_ids_2 = np.digitize(links2[2], bins2)
    contacts = np.zeros([len(bins1), len(bins2)])
    for i in np.unique(bin_ids_1):
        for j in np.unique(bin_ids_2):
            loop_1 = links1[pd.Series(bin_ids_1)==i]
            loop_2 = links2[pd.Series(bin_ids_2)==j]
            reads_1 = np.unique(loop_1[3])
            reads_2 = np.unique(loop_2[3])
            contacts_ij = set(reads_1) & set(reads_2)
            contacts[i-1][j-1]+=len(contacts_ij)
    c = pd.DataFrame(contacts)
    return c

def unphasedTransMap(graph_path, chrom1, chrom2):
    links_1 = pd.read_csv(graph_path+chrom1+'.dat', 
                        sep='\t', header=None).sort_values(2)
    links_2 = pd.read_csv(graph_path+chrom2+'.dat',
                      sep='\t', header=None).sort_values(2)
    trans_links = set(links_1[3]) & set(links_2[3])
    links_1 = links_1[links_1[3].isin(trans_links)].reset_index(drop=True)
    links_2 = links_2[links_2[3].isin(trans_links)].reset_index(drop=True)
    mat_dict = {}
    counter = 0
    mat_dict = makeContact(links_1, links_2, chrom1, chrom2)    
    return mat_dict

def phaseGraph(graph, hap):
    graph = graph.copy()
    getGraphGT = lambda x: [2*int(i.split('_')[1]==i.split('_')[2])-1 for i in x[0].values]
    gt = getGraphGT(graph)
    graph[5] = gt
    overlap = lambda hap,links: set(hap[1]) & set(links[2])
    graph = graph[graph[2].isin(overlap(hap, graph))]
    make_hap_dict = lambda x: {i:j for i,j in zip(x[1], x[4])} ##Change this! (hap may be in x[5] instead of x[4])
    hap_dict = make_hap_dict(hap)
    hap_col = []
    for j in range(graph.shape[0]):
        pos = graph[2].values[j]
        hap = hap_dict[pos]
        hap_col.append(hap)
    graph[6] = hap_col
    graph[7] = graph[5].values*graph[6].values
    return graph.reset_index(drop=True)

def phasedTransMap(chrom1, chrom2, graph_file, hap_path, binsize):
    links_1 = pd.read_csv(graph_file+chrom1[3:]+'.dat',sep='\t', header=None).sort_values(2)
    links_2 = pd.read_csv(graph_file+chrom2[3:]+'.dat',sep='\t', header=None).sort_values(2)
    trans_links = set(links_1[3]) & set(links_2[3])
    links_1 = links_1[links_1[3].isin(trans_links)].reset_index(drop=True)
    links_2 = links_2[links_2[3].isin(trans_links)].reset_index(drop=True)

    hap_1 = pd.read_csv(hap_path+chrom1[3:]+'.dat',
                        header=None, sep="\t")
    hap_2 = pd.read_csv(hap_path+chrom2[3:]+'.dat',
                        header=None, sep="\t")

    hap_1 = hap_1[hap_1[4]!=0] ##Change this! (hap may be in x[5] instead of x[4])
    hap_2 = hap_2[hap_2[4]!=0] ##Change this! (hap may be in x[5] instead of x[4])

    links_1 = phaseGraph(links_1, hap_1)
    links_2 = phaseGraph(links_2, hap_2)

    mat_dict = {}
    counter = 0

    for i in [1,-1]:
        for j in [1,-1]:
            loop_hap_1 = links_1[links_1[7]==i].copy().reset_index(drop=True)
            loop_hap_2 = links_2[links_2[7]==j].copy().reset_index(drop=True)
            print(chrom1,' hap ', i, chrom2,' hap ', j)
            mat_dict[counter] = makeContact(loop_hap_1, loop_hap_2, chrom1, chrom2, binsize)    
            counter+=1
    return mat_dict

def phasedIntraLinks(graph_path, hap_path, chrom):
    links = pd.read_csv(graph_path+chrom+'.dat', sep='\t', header=None).sort_values(2).reset_index(drop=True)
    hap = pd.read_csv(hap_path+chrom+'.dat',
                        header=None, sep="\t")
    hap = hap[hap[4]!=0]
    phased_graph = phaseGraph(links, hap)
    contact_dict = {}
    counter = 0
    hic_bins = range(0, getSize(chrom), int(5e5))
    contacts = np.zeros([len(hic_bins), len(hic_bins)])
    for i in [1,-1]:
        loop_graph = phased_graph.copy()[phased_graph[7]==i].reset_index(drop=True)
        bins_ids = np.digitize(loop_graph[2], bins=hic_bins)
        for j in range(np.unique(bins_ids).shape[0]):
            for k in range(j, np.unique(bins_ids).shape[0]):
                loop_1 = loop_graph[pd.Series(bins_ids)==np.unique(bins_ids)[j]]
                loop_2 = loop_graph[pd.Series(bins_ids)==np.unique(bins_ids)[k]]
                reads_1 = np.unique(loop_1[3])
                reads_2 = np.unique(loop_2[3])
                contacts_ij = set(reads_1) & set(reads_2)
                if i == 1:
                    contacts[j][k]+=len(contacts_ij)
                else:
                    contacts[k][j]+=len(contacts_ij)

    plt.rcParams['figure.figsize'] = [4, 4]
    fig,ax = plt.subplots()
    sns.heatmap(pd.DataFrame(contacts), cmap='Reds', vmax=5, ax=ax)
    plt.show()
```

```python
def plotPhasedInter(mat, cmax, chrom1, chrom2, fname):
    
    plt.rcParams['figure.figsize'] = [8.5, 8]
    loop_bands1 = chr_bands[chr_bands['chrom']==chrom1]
    loop_bands2 = chr_bands[chr_bands['chrom']==chrom2]
    
    fig, axs = plt.subplots(3, 3, gridspec_kw={'height_ratios': [5, 5, 0.275], 'width_ratios': [0.25, 5, 5]})
    for i in range(2):
        axs[i,0].set_xlim([-1,0])
        axs[i,0].set_ylim([getSize(chrom1),0])
        axs[i,0].spines["top"].set_visible(False)
        axs[i,0].spines["bottom"].set_visible(False)
        axs[i,0].spines["right"].set_visible(False)
        axs[i,0].set_xticks([])

    for j in range(loop_bands1.shape[0]):
                row = loop_bands1.iloc[j]
                rect=mpatches.Rectangle((-0.2, row['chromStart']),-0.6,row['chromEnd']-row['chromStart'], fill = True,edgecolor = 'black',facecolor = row['color'],linewidth = 1)
                axs[0,0].add_patch(rect)
    for j in range(loop_bands1.shape[0]):
                row = loop_bands1.iloc[j]
                rect=mpatches.Rectangle((-0.2, row['chromStart']),-0.6,row['chromEnd']-row['chromStart'], fill = True,edgecolor = 'black',facecolor = row['color'],linewidth = 1)
                axs[1,0].add_patch(rect)
    
    sns.heatmap(mat[0], vmin=0, vmax=cmax, cmap='Reds', ax = axs[0,1], cbar=False, 
               xticklabels=False, yticklabels=False)
    sns.heatmap(mat[1], vmin=0, vmax=cmax, cmap='Reds', ax = axs[0,2], cbar=False, 
               xticklabels=False, yticklabels=False)
    sns.heatmap(mat[2], vmin=0, vmax=cmax, cmap='Reds', ax = axs[1,1], cbar=False, 
               xticklabels=False, yticklabels=False)
    sns.heatmap(mat[3], vmin=0, vmax=cmax, cmap='Reds', ax = axs[1,2], cbar=False, 
               xticklabels=False, yticklabels=False)
    for i in range(2):
        axs[i,1].axhline(y=0, color='k',linewidth=1)
        axs[i,1].axhline(y=mat[0].shape[0], color='k',linewidth=1)
        axs[i,1].axvline(x=0, color='k',linewidth=1)
        axs[i,1].axvline(x=mat[0].shape[1], color='k',linewidth=1)
        axs[i,2].axhline(y=0, color='k',linewidth=1)
        axs[i,2].axhline(y=mat[0].shape[0], color='k',linewidth=1)
        axs[i,2].axvline(x=0, color='k',linewidth=1)
        axs[i,2].axvline(x=mat[0].shape[1], color='k',linewidth=1)

    axs[2,0].remove()
    
    for i in range(1,3):
        axs[2,i].set_xlim([0, getSize(chrom2)])
        axs[2,i].set_ylim([-1,0])
        axs[2,i].spines["top"].set_visible(False)
        axs[2,i].spines["left"].set_visible(False)
        axs[2,i].spines["right"].set_visible(False)
        axs[2,i].set_yticks([])
        
    for j in range(loop_bands2.shape[0]):
                row = loop_bands2.iloc[j]
                rect=mpatches.Rectangle((row['chromStart'],-0.2),row['chromEnd']-row['chromStart'],-0.6, fill = True,edgecolor = 'black',facecolor = row['color'],linewidth = 1)
                axs[2,1].add_patch(rect)
    for j in range(loop_bands2.shape[0]):
                row = loop_bands2.iloc[j]
                rect=mpatches.Rectangle((row['chromStart'],-0.2),row['chromEnd']-row['chromStart'],-0.6, fill = True,edgecolor = 'black',facecolor = row['color'],linewidth = 1)
                axs[2,2].add_patch(rect)
    #plt.tight_layout()
    plt.savefig(fname, transparent=True, format='pdf')
```
